```python
#converting the timestamp into timedate data type
def time_convert(value):
    to_sec=value/1000
    to_date=datetime.fromtimestamp(to_sec)
    return to_date.strftime("%A at %H:%M:%S (%d%B%Y)")

df["Event_occured"]=df["Event_occured"].apply(lambda x: time_convert(x))
```

```python
#changing the is_tsunami 0 & 1 value to TRUE and False Value

df['is_Tsunami']=df['is_Tsunami'].apply(lambda x:True if x==1 else False)
```

```python
#determining the type of earthquake based on Magnitude Value Measurement
def impact_class(input):

    if input >= 0.00 and input <=1.99:
        return "minor"
    if input >= 2.00 and input <=2.99:
        return "minor"
    if input >=3.00 and input <=3.99:
        return "Light"
    if input >= 4.00 and input <=5.99:
        return "Strong"
    if input >= 6.00 and input <=10.00:
        return "Great Imapact"

df["Earthquake_impact"]=df["Magnitude"].apply(lambda x:impact_class(x))
```

```python
# Load the preprocessed data into a Pandas DataFrame
data = pd.read_csv('USGS_earthquake_dataset.csv')

# Select the features and target variable
features = data[['Magnitude', 'Depth', 'Latitude', 'logitude']]
target = data['Earthquake_impact']

# Create a Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=200, random_state=42)

# Define the StratifiedKFold object for cross-validation
# We use StratifiedKFold to maintain the proportion of classes in each fold
k_folds = 5
skf = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=42)

# Define the few-shot size
few_shot_size = 15  # Adjust this value as needed

# Initialize a list to store the accuracy scores for each fold
accuracy_scores = []

# Perform few-shot cross-validation
for train_index, test_index in skf.split(features, target):
    # Create few-shot training data using a small subset of the training set
    few_shot_data = features.iloc[train_index].head(few_shot_size)
    few_shot_labels = target.iloc[train_index].head(few_shot_size)

    # Separate test data
    test_data = features.iloc[test_index]
    test_labels = target.iloc[test_index]

    # Train the Random Forest model on the few-shot training data
    rf_classifier.fit(few_shot_data, few_shot_labels)

    # Evaluate the model on the test set
    test_accuracy = rf_classifier.score(test_data, test_labels)

    # Store the test accuracy for this fold
    accuracy_scores.append(test_accuracy)

# Calculate and print the average accuracy across all folds
average_accuracy = sum(accuracy_scores) / k_folds
print(f"Average few-shot cross-validation accuracy: {average_accuracy:.2f}")
```

```python
# Load the earthquake dataset
data = pd.read_csv('USGS_earthquake_dataset.csv')

# Preprocess the data
features = ['Magnitude', 'Depth', 'Latitude', 'logitude']
target = 'Earthquake_impact'

# Standardize the features
scaler = StandardScaler()
data[features] = scaler.fit_transform(data[features])

# Convert the target labels to integer values
data[target], _ = pd.factorize(data[target])
```

```python
# Define a simple neural network model for classification
class SimpleNN(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(64, output_dim)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```

```python
# Instantiate the model
input_dim = len(features)
output_dim = len(data[target].unique())
model = SimpleNN(input_dim, output_dim)

# Define an optimizer
optimizer = Adam(model.parameters(), lr=0.001)

# Define a loss function
loss_fn = nn.CrossEntropyLoss()
```

```python
def train_maml(model, data, num_iterations=100, inner_lr=0.01, outer_lr=0.001):
    outer_optimizer = Adam(model.parameters(), lr=outer_lr)

    # Check if data is a list
    if not isinstance(data, list):
        raise ValueError("Data should be a list of tasks")

    # Iterate through iterations
    for iteration in range(num_iterations):
        # Iterate through tasks
        for task_idx, task_data in enumerate(data):
            # Ensure the task data is a DataFrame
            if not isinstance(task_data, pd.DataFrame):
                raise ValueError(f"Task {task_idx} is not a DataFrame")

            # Check if task data contains required columns
            if not all(col in task_data.columns for col in features + [target]):
                raise ValueError(f"Task {task_idx} data does not contain required columns: {features} and {target}")

            # Split the task data into support set and query set
            support_set = task_data.sample(frac=0.8, random_state=42)
            query_set = task_data.drop(support_set.index)

            # Create support and query inputs and targets
            support_inputs = torch.tensor(support_set[features].values, dtype=torch.float32)
            support_targets = torch.tensor(support_set[target].values, dtype=torch.int64)
            query_inputs = torch.tensor(query_set[features].values, dtype=torch.float32)
            query_targets = torch.tensor(query_set[target].values, dtype=torch.int64)

            # Use inner loop context with higher
            with innerloop_ctx(model, optimizer, copy_initial_weights=True) as (fmodel, diffopt):
                # Perform a few inner loop steps on the support set
                for _ in range(1):  # Adjust the number of inner loop steps
                    support_preds = fmodel(support_inputs)
                    support_loss = loss_fn(support_preds, support_targets)
                    diffopt.step(support_loss)

                # Evaluate on the query set
                query_preds = fmodel(query_inputs)
                query_loss = loss_fn(query_preds, query_targets)

            # Update the meta-model using outer optimizer
            outer_optimizer.zero_grad()
            query_loss.backward()
            outer_optimizer.step()

        print(f"Iteration {iteration + 1}/{num_iterations}, Query Loss: {query_loss.item():.4f}")
```

```python
import numpy as np

class PrototypeNetwork:
    def __init__(self, features, target):
        self.features = features
        self.target = target
        self.prototypes = {}

    def fit(self, X_train, y_train):
        # Group data points by class
        grouped_data = X_train.groupby(y_train)

        # Calculate prototypes for each class
        for label, group in grouped_data:
            self.prototypes[label] = group.mean()

    def predict(self, X_test):
        # Calculate distances between query points and prototypes
        distances = pd.DataFrame(index=X_test.index, columns=self.prototypes.keys())
        for label, prototype in self.prototypes.items():
            distances[label] = np.linalg.norm(X_test[self.features].values - prototype.values, axis=1)

        # Assign labels based on nearest prototype
        return distances.idxmin(axis=1)

# Initialize and train the prototype network
prototype_net = PrototypeNetwork(features, target)
prototype_net.fit(X_train, y_train)

# Predict on the test set
predictions = prototype_net.predict(X_test)

# Evaluate the predictions
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```

```python
# Section 1: Import Libraries
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score

# Section 2: Load Dataset
data = pd.read_csv("USGS_earthquake_dataset.csv")

# Section 3: Define Features and Target
features = ['Magnitude', 'Depth', 'Latitude', 'logitude']  # Corrected 'logitude' to 'Longitude'
target = 'Earthquake_impact'

# Check if the dataset loaded properly
if data is not None:
    # Print the first few rows to see if data is loaded correctly
    print(data.head())
    # Check the shape of the DataFrame
    print(data.shape)

    # Prepare Data for Classification

    # Standardize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(data[features])

    # Convert target labels to integer values
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(data[target])

    # Define KFold Cross-Validation
    kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    # Define Neural Network Model for Classification
    class SimpleModel(nn.Module):
        def __init__(self, input_size, output_size):
            super(SimpleModel, self).__init__()
            self.fc1 = nn.Linear(input_size, 64)
            self.fc2 = nn.Linear(64, output_size)

        def forward(self, x):
            x = torch.relu(self.fc1(x))
            x = self.fc2(x)
            return x

    # Perform Cross-Validation for Classification
    cv_scores = []

    for train_index, test_index in kfold.split(X_scaled, y_encoded):
        X_train_cv, X_test_cv = X_scaled[train_index], X_scaled[test_index]
        y_train_cv, y_test_cv = y_encoded[train_index], y_encoded[test_index]

        # Train the model (Classification)
        model = SimpleModel(input_size=X_train_cv.shape[1], output_size=len(np.unique(y_encoded)))
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(model.parameters(), lr=0.001)

        for epoch in range(100):
            model.train()
            optimizer.zero_grad()
            outputs = model(torch.tensor(X_train_cv, dtype=torch.float32))
            loss = criterion(outputs, torch.tensor(y_train_cv, dtype=torch.long))
            loss.backward()
            optimizer.step()

        # Evaluate the model (Classification)
        model.eval()
        with torch.no_grad():
            outputs = model(torch.tensor(X_test_cv, dtype=torch.float32))
            _, predicted = torch.max(outputs, 1)
            accuracy = accuracy_score(y_test_cv, predicted.numpy())
            cv_scores.append(accuracy)

    # Print Mean Cross-Validation Accuracy for Classification
    print("Mean Cross-Validation Accuracy for Classification:", np.mean(cv_scores))

    # Meta Learning with Few-shot Learning

    # Define few-shot learning task size
    task_size = 20

    # Define the number of meta-tasks
    num_meta_tasks = 5

    # Function to create tasks for meta-learning
    def create_meta_tasks(data, num_tasks, task_size):
        meta_X_train = []
        meta_X_test = []
        meta_y_train = []
        meta_y_test = []

        for _ in range(num_tasks):
            task_data = data.sample(task_size)
            X_task = task_data[features]
            y_task = label_encoder.transform(task_data[target])
            X_task_train, X_task_test, y_task_train, y_task_test = train_test_split(X_task, y_task, test_size=0.2, random_state=42)
            meta_X_train.append(X_task_train)
            meta_X_test.append(X_task_test)
            meta_y_train.append(y_task_train)
            meta_y_test.append(y_task_test)

        return meta_X_train, meta_X_test, meta_y_train, meta_y_test

    # Create meta-tasks
    meta_X_train, meta_X_test, meta_y_train, meta_y_test = create_meta_tasks(data, num_meta_tasks, task_size)

    # Train and Evaluate Meta-Learning Model with Few-shot Learning

    # Initialize lists to store meta-learning accuracy scores
    meta_learning_scores = []

    # Meta-learning loop
    for meta_task_index in range(num_meta_tasks):
        # Initialize model for meta-learning
        meta_model = SimpleModel(input_size=X_train_cv.shape[1], output_size=len(np.unique(y_encoded)))
        meta_optimizer = optim.Adam(meta_model.parameters(), lr=0.001)

        # Few-shot learning on the current meta-task
        for epoch in range(5):  # Few-shot learning epoch
            # Train the meta-model on the meta-task
            meta_model.train()
            meta_optimizer.zero_grad()

            # Convert meta-task data to NumPy array
            meta_X_train_np = np.array(meta_X_train[meta_task_index], dtype=np.float32)
            meta_y_train_np = np.array(meta_y_train[meta_task_index], dtype=np.int64)


            # Pass NumPy arrays to the model
            meta_outputs = meta_model(torch.tensor(meta_X_train_np))
            meta_loss = criterion(meta_outputs, torch.tensor(meta_y_train_np))
            meta_loss.backward()
            meta_optimizer.step()

        # Evaluate the meta-model on the meta-task test set
        meta_model.eval()
        with torch.no_grad():
            meta_outputs = meta_model(torch.tensor(meta_X_test[meta_task_index].values, dtype=torch.float32))
            _, meta_predicted = torch.max(meta_outputs, 1)
            meta_accuracy = accuracy_score(meta_y_test[meta_task_index], meta_predicted.numpy())
            meta_learning_scores.append(meta_accuracy)

    # Print Mean Meta-Learning Accuracy with Few-shot Learning
    print("Mean Meta-Learning Accuracy with Few-shot Learning:", np.mean(meta_learning_scores))

else:
    print("Error: Dataset not loaded properly.")
```
